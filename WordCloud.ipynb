```python
from wordcloud import WordCloud, STOPWORDS

import matplotlib.pyplot as plt
import numpy as np

uninteressant = "und von Der das den wir ist die auf im Auf oder zum zu ich mich zu auf mit bin"
liste_der_unerwuenschten_woerter = uninteressant.split()
STOPWORDS.update(liste_der_unerwuenschten_woerter)

# Maske fr WordCloud erstellen
x, y = np.ogrid[:1500, :1500]
mask = (x - 900) ** 2 + (y - 900) ** 2 > 700 ** 2
mask = 255 * mask.astype(int)


text = """Hallo! Ich bin Brbel.

Ich liebe Sprche, die mich zum Nachdenken, Schmunzeln oder Lachen bringen.

Auf Stoff gestickt (und zu Gebrauchsgegenstnden genht) begleiten sie mich auf meiner Tour durch den Tag, erinnern mich an das Wesentliche und die Schnheit im Leben, fhren mich zu Dankbarkeit und Achtsamkeit oder bringen mich zum Lcheln.

Wenn ich nicht gerade hinter der Nhmaschine sitze, findest du mich

beim Spazierengehen in der Natur
bei guten Freunden und gutem Essen
beim Lesen auf dem Sofa
Ich lebe in Frstenfeldbruck, bin verheiratet und habe 2 Shne.

Auf Tour bin ich gern an der Ostsee, trinke am liebsten Latte Macchiato und finde alles Schne ganz wunderbar! 
Hier bei Fadentour findest du mit Liebe und Sorgfalt handgemachte Kulturbeutel, Kissen, Babydecken, Beulentrster & einiges mehr.

Vieles davon besticke ich mit Wrtern oder Sprchen.

Meine Stoffe kaufe ich meistens in kleinen Mengen, deshalb gibt es hchstens Kleinstauflagen zu erwerben.

Auf Wunsch fertige ich fr dich dein ganz persnliches Lieblingsstck mit einem Namen oder einem Spruch deiner Wahl."""

# Generate a word cloud image
wordcloud = WordCloud().generate(text)
wordcloud = WordCloud(background_color="white",width=1920, height=1080, mask=mask).generate(text)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
