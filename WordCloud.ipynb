```python
from wordcloud import WordCloud, STOPWORDS

import matplotlib.pyplot as plt
import numpy as np

uninteressant = "und von Der das den wir ist die auf im Auf oder zum zu ich mich zu auf mit bin"
liste_der_unerwuenschten_woerter = uninteressant.split()
STOPWORDS.update(liste_der_unerwuenschten_woerter)

# Maske für WordCloud erstellen
x, y = np.ogrid[:1500, :1500]
mask = (x - 900) ** 2 + (y - 900) ** 2 > 700 ** 2
mask = 255 * mask.astype(int)


text = """Hallo! Ich bin Bärbel.

Ich liebe Sprüche, die mich zum Nachdenken, Schmunzeln oder Lachen bringen.

Auf Stoff gestickt (und zu Gebrauchsgegenständen genäht) begleiten sie mich auf meiner Tour durch den Tag, erinnern mich an das Wesentliche und die Schönheit im Leben, führen mich zu Dankbarkeit und Achtsamkeit oder bringen mich zum Lächeln.

Wenn ich nicht gerade hinter der Nähmaschine sitze, findest du mich

beim Spazierengehen in der Natur
bei guten Freunden und gutem Essen
beim Lesen auf dem Sofa
Ich lebe in Fürstenfeldbruck, bin verheiratet und habe 2 Söhne.

Auf Tour bin ich gern an der Ostsee, trinke am liebsten Latte Macchiato und finde alles Schöne ganz wunderbar! 
Hier bei Fadentour findest du mit Liebe und Sorgfalt handgemachte Kulturbeutel, Kissen, Babydecken, Beulentröster & einiges mehr.

Vieles davon besticke ich mit Wörtern oder Sprüchen.

Meine Stoffe kaufe ich meistens in kleinen Mengen, deshalb gibt es höchstens Kleinstauflagen zu erwerben.

Auf Wunsch fertige ich für dich dein ganz persönliches Lieblingsstück mit einem Namen oder einem Spruch deiner Wahl."""

# Generate a word cloud image
wordcloud = WordCloud().generate(text)
wordcloud = WordCloud(background_color="white",width=1920, height=1080, mask=mask).generate(text)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
